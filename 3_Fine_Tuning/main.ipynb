```python
import torch

PROMPT = "image of cloudsss"
BATCH_SIZE = 4 # Так же овечает за seed. Если меняете, не забудте поменять количество в make_image_grid
generator = [torch.Generator("cuda").manual_seed(i) for i in range(BATCH_SIZE)]
def get_inputs(batch_size=BATCH_SIZE):
    generator = [torch.Generator("cuda").manual_seed(i) for i in range(batch_size)]
    prompts = batch_size * [PROMPT]
    num_inference_steps = 20

    return {"prompt": prompts, "generator": generator, "num_inference_steps": num_inference_steps}
```

Generation of images

```python
from diffusers import DiffusionPipeline

modelPath = r"Z:\MLL_HG"
pipelineTrained = DiffusionPipeline.from_pretrained(modelPath,
                                            torch_dtype=torch.float16,
                                            device_map="cuda")
images = pipelineTrained(**get_inputs()).images
torch.cuda.empty_cache()
```

```python

pipelineUn = DiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5",
                                            torch_dtype=torch.float16,
                                            device_map="cuda",
                                            cache_dir="Z:\MLL_HG")
imagesUnt = pipelineUn(**get_inputs()).images
torch.cuda.empty_cache()
```

```python
from diffusers.utils import make_image_grid 

make_image_grid(imagesUnt, rows=4, cols=1).save("pics1.jpg")
```
