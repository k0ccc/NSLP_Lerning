Init pipeline

```python

from diffusers import DiffusionPipeline
import torch


torch.cuda.empty_cache()

pipeline = DiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5",
                                            torch_dtype=torch.float16,
                                            device_map="cuda",
                                            cache_dir="Z:\MLL_HG")
```

helpful func and const

```python
PROMPT = "texture of lava for 3D UV"
BATCH_SIZE = 8 # Так же овечает за seed. Если меняете, не забудте поменять количество в make_image_grid


def get_inputs(batch_size=BATCH_SIZE):
    generator = [torch.Generator("cuda").manual_seed(i) for i in range(batch_size)]
    prompts = batch_size * [PROMPT]
    num_inference_steps = 20

    return {"prompt": prompts, "generator": generator, "num_inference_steps": num_inference_steps}
```

Generation of images

```python
from diffusers.utils import make_image_grid 


generator = [torch.Generator("cuda").manual_seed(i) for i in range(BATCH_SIZE)]
images = pipeline(**get_inputs()).images
make_image_grid(images, rows=2, cols=4).save("8_pic.jpg")
```
